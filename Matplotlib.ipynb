```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([0, 10])
y = np.array([0, 100])

print(np.__version__)
print(x)
print(y)
```

```python
plt.plot(x, y)
plt.show()
```

```python
x = np.array([1, 8])
y = np.array([3, 10])

plt.plot(x, y)
plt.show()
```

```python
x = np.array([1, 8])
y = np.array([3, 10])

plt.plot(x, y, 'o')
plt.show()
```

```python
x = np.array([4, 5, 9, 15])
# plt.plot(x, marker = 'o')
plt.plot(x, marker = '*')
plt.show()
```

```python
x = np.array([4, 5, 9, 15])
plt.plot(x, 'o-.g')
plt.show()
```

```python
x = np.array([9, 5, 20, 60])
plt.plot(x, 'o-.y', ms = 20, mec = 'r')
plt.show()
```

```python
x = np.array([9, 5, 20, 60])
plt.plot(x,'o', linestyle='dotted')
plt.show()
```

```python
x = np.array([9, 5, 20, 60])
plt.plot(x, color='pink', linewidth='15')
plt.show()
```

```python
x = np.array([0, 60, 100])
y = np.array([0, 30, 80, 100])

plt.plot(x, color="red")
plt.plot(y, color="green")

plt.show()
```

```python
x = np.array([0, 60, 100, 50])
y = np.array([0, 30, 80, 100])
z = np.array([10, 20, 40, 90])

plt.plot(x, y, z)
plt.show()
```

```python
x = np.array([30, 51, 64, 37, 28])
y = np.array([20, 41, 54, 57, 28])

plt.plot(x, y)

plt.title('Student Mark Sheet', loc='right')
plt.xlabel('X Label', loc='right')
plt.ylabel('Y Label', loc="top")

plt.grid(axis = 'x', color="green", linestyle='--')


plt.show()
```

```python
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 2)
plt.plot(x,y)

plt.title('Title')
plt.suptitle('Super Title')

plt.show()
```

```python
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.scatter(x, y, color='red')
plt.show()
```

```python
x = np.array(['A', 'B', 'C', 'D'])
y = np.array([3, 8, 1, 10])

plt.scatter(x, y)
plt.show()
```

```python
x = np.array(['Apple', 'Mango', 'Orang', 'Banana'])
y = np.array([50, 80, 30, 50])

plt.bar(x, y, width=0.4)
plt.scatter(x, y, color='blue')

plt.title("Fruit")
plt.xlabel('Fruit Name')
plt.ylabel('Fruit Prices')
plt.grid(color='blue', linestyle='--')
plt.show()
```

```python
x = np.array(['Apple', 'Mango', 'Orang', 'Banana'])
y = np.array([50, 80, 30, 50])

plt.barh(x, y)

plt.title("Fruit")
plt.xlabel('Fruit Prices')
plt.ylabel('Fruit Name')
plt.grid(axis = 'x', color='blue')
plt.show()
```

```python
x = np.array(['Apple', 'Mango', 'Orang', 'Banana'])
y = np.array([50, 80, 30, 50])

plt.pie(y, labels = x)

plt.title("Fruit")
plt.show()
```
